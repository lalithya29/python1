```python
vehicle=input("choose:\n1.bike\n2.auto\n3.car\n") 
if vehicle=="1": 
   kmph=int(input("enter the distance to travel on bike:")) 
   if kmph>0: 
     if kmph<=5: 
        print(kmph*5) 
                                                                                  
     elif kmph>=6 and kmph<=12: 
        print(5*5+(kmph-5)*6) 
     elif kmph>=13 and kmph<=20: 
        print(5*5+42+(kmph-12)*8) 
     elif kmph>=21 and kmph<=30: 
        print(5*5+42+64+(kmph-20)*10) 
     else: 
        print("ride not available") 
   else: 
     print("distance should be minimum 1 to 5") 
elif vehicle=="2": 
    kmph=int(input("enter the distance to travel on auto:")) 
    if kmph>0: 
     if kmph<=9: 
        print(kmph*10) 
     elif kmph>=10 and kmph<=20: 
        print(90+(kmph-9)*15) 
     elif kmph>=21 and kmph<=30: 
        print(90+165+(kmph-20)*20) 
     elif kmph>=31 and kmph<=50: 
        print(90+165+200+(kmph-30)*22) 
     else: 
        print("ride not available") 
    else: 
        print("distance should be minimum 1 to 9") 
                                                                                  
elif vehicle=="3": 
    kmph=int(input("enter the distance to travel on car:")) 
    food=int(input()) 
    if kmph>0: 
     if kmph<=50: 
        print((kmph*17)+200) 
     elif kmph>=51 and kmph<=100: 
        print(850+((kmph-50)*21)+300+food) 
     elif kmph>=101 and kmph<=160: 
        print(850+(50*21)+((kmph-100)*30)+300+food) 
     elif kmph>=161 and kmph<=200: 
        print(850+(50*21)+(60*30)+((kmph-160)*33)+300+food) 
     else: 
        print("ride not available") 
    else: 
        print("distance should be minimum 1 to 50") 
else: 
    print("exit") 
```
