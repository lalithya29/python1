```python

def Fact(n):
    if n==1:
       return 1
    else:
       return n*Fact(n-1)       
print(Fact(5))
```

```python
def fib(n):
    if n==0:
       return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
print(fib(6))
```

```python
def fib(n):
    a=0;b=1;s=0
    for i in range(n):
        s+=a
        a,b=b,a+b  
    return s    
fib(6)    
```
