```python
import random
n=random.randrange(1,100)
for i in range(3):#3 chances for user to guess number
    print(f"chance{i+1}")
    num=int(input("enter the guess number"))
    if num==n:
       print("you won")
       print("The number is",n)
       break #if user won it breaks
    elif num<n:
       print("lower!try again")
    elif num>n:
       print("higher!try again")
else:
    print("computer won")
    print("The number is",n)
```
