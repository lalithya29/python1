```python
#Set Methods 
s={1} 
print(s) 
s={1,True} 
print(s) 
s={False,0} 
print(s) 
s={1,2,3,4,True,"str",False}#it takes true=1 and so it returns {1,2,3,4,str,false] 
print(s) 
s.add(0)#it will not add o because set dont allow duplicate values 
print(s) 
s.update({False,True,100}) 
print(s) 
print(s.pop())#it will remove arbitary element since set is unordered 
print(s) 
s.discard("str")#it removes if element present if not it will not arise any error 
print(s) 
s.clear()#it removes entire element in set 
print(s) 
s={1,2,3} 
s.update([4,5])#it returns entire elements as set because list is mutable 
print(s)
```

```python
s=[1,2,3] 
s.append({4,5,6})#it returns [1,2,3{4,5,6}]since set is immutable 
print(s) 
s.extend({99,10})#it returns all elements as set in extend 
print(s) 
s={1,2,3,(4,5,6)} 
print(s) 
s=set(map(int,input().split())) 
print(s) 
print(min(s)) 
print(max(s)) 
print(sum(s))
```

```python
kohli={"pardhu","manikanta","sai","john"} 
dhoni={"pardhu","b","c"} 
#union ---> all 
print(kohli.union(dhoni))#it returns all values of both sets 
print(kohli | dhoni) 
# #interscetion:  --- > common 
print(kohli.intersection(dhoni))#it returns common values in both set 
print(dhoni & kohli) 
#difference: 
print(kohli.difference(dhoni))#it returns no common values in 1st set
print(dhoni-kohli) 
#  
# #symmetric_difference: 
print(kohli.symmetric_difference(dhoni))#it returns no common values from both the set 
print(dhoni^kohli)
```

```python
a={1,2,3,4,5} 
b={1,2,3,4} 
#issubset:<= 
print(a.issubset(b))# it returns true if all element in set present in anotherset 
print(b.issubset(a)) 
print(a<=b) 
#issuperset:>= 
print(a.issuperset(b))#it returns true if all element of another set present in set   
print(b>=a) 
#isdisjoin:#it returns true if no elements in set matched 
print(a.isdisjoint(b))
```
