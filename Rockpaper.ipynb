```python
import random
c=random.choice(["Rock","Paper","Scissors"])
playerscore=0
computerscore=0
for i in range(5):#Rounds
    print(f"roundno{i+1}")
    p=input("enter Rock or paper or scissors:").capitalize()
    if p==c:
       print(f"Both player{p} and computer draw{c}")
    elif (p=="Paper"and c=="Rock")or (p=="Rock" and c=="Scissors") or (p=="Scissors" and c=="Paper"):
       print("player wins")
       print(f"players choice {p} and computer choice {c}")
       playerscore+=1
       print(f"player score {playerscore}")
    else:
        print("computer wins")
        print(f"players choice {p} and computer choice {c}")
        computerscore+=1
        print(f"computer score {computerscore}")
if playerscore>computerscore:
    print(f"player wins by {playerscore} and computer losses by {computerscore}")
elif playerscore<computerscore:
    print(f"computer wins by {computerscore} and player losses by {playerscore}")
else:
    print(f"both draw player score {playerscore} and computer score {computerscore}")
```
