```python
#print numbers from 1 to 10
for i in range(1,10):
    print(i)
```

```python
#print numbers from 10 to 1
i=11
while i>1:
    i-=1
    print(i)
```

```python
#print all even numbers from 1 to 20
for i in range(1,20):
    if i%2==0:
       print(i)
#or
i=-2
while i<18:
    i+=2
    print(i)
```

```python
#print the multiplication table of a given number
n=input()
m=int(n)
for i in range(1,11):
    print(f"{n}*{i}=",m*i)
#or
n=input()
m=int(n)
i=0
while i<10:
    i+=1
    print(f"{n}*{i}=",m*i)
```

```python
#print the factorial of given number
n=int(input())
fact=1
for i in range(n):
    k=n-i
    fact*=k
print(fact)
#or
n=int(input())
fact=1
i=0
while i<n:
    k=n-i
    fact*=k
    i+=1
print(fact)
```

```python
#print the digit sum
n=input()
sum=0
for i in n:
    sum+=int(i)  
print(sum)
```

```python
#print 1 to 50 odd numbers
for i in range(50):
    if i%2!=0:
       print(i)
```

```python
#print the largest number in the list
n=list(map(int,input().split()))
for i in n:
    count=0
    for j in n:
     if i>=j:
        count+=1
    if count==len(n):
      print(i)
      break
```
